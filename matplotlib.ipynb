```
import matplotlib.pyplot as plt

#basic plot

x = [6,7,9]
y = [8,7,9]

x1 = [10,11,12]
y1 = [13,14,15]

plt.plot(x,y, label='line1')
plt.plot(x1,y1, label = 'line2')

#Giving Labels to axis

plt.xlabel("Vertical Axis")
plt.ylabel("horizontal Axis")

#title

plt.title("Basic Plot\n(check it out)")
plt.legend()

plt.show()
```

```
x = [2,4,7,5]
y = [12,3,20,11]

#making a bar graph

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bar graph')
plt.bar(x,y)                    #to plot a bar graph
plt.show()
```
